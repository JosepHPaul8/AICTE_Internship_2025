```python
#Load Datasets
dataset_root = "/kaggle/input/the-wildfire-dataset/the_wildfire_dataset_2n_version"
train_dir = os.path.join(dataset_root, "train")
val_dir = os.path.join(dataset_root, "val")
test_dir = os.path.join(dataset_root, "test")
```

```python
#list Class
classes = os.listdir(train_dir)
numclasses=len(classes)
print(f"Total Classes: {numclasses}")
print(f"Classes: {classes}")
```

```python
plt.figure(figsize=(12, 10))

for j, class_label in enumerate(classes[:2]):  # Iterate over the first two classes
    class_path = os.path.join(train_dir, class_label)

    for i in range(5):
        img_name = os.listdir(class_path)[i]
        img_path = os.path.join(class_path, img_name)
        img = plt.imread(img_path)

        plt.subplot(2, 5, j * 5 + i + 1)  # Arrange images in 2 rows
        plt.imshow(img)
        plt.title(f'{class_label} \n shape: {img.shape}')
        plt.axis("off")

plt.show()
```
